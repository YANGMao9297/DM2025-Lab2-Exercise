### 1.1 Preprocessing Steps

在本次情緒分類競賽中，我對文本資料進行了以下預處理步驟：

1. **轉換為小寫**：將所有文字統一轉換為小寫，確保一致性。

2. **移除 URL**：使用正則表達式移除 `http://`、`https://`、`www` 開頭的連結。

3. **移除 @mentions**：移除社群媒體中常見的 `@用戶名` 標記。

4. **處理 Hashtags**：移除 `#` 符號但保留 hashtag 的文字內容。

5. **移除 HTML 標籤**：清除任何殘留的 HTML 標籤。

6. **移除特殊字符與數字**：只保留英文字母和空白。

7. **分詞 (Tokenization)**：使用 NLTK 的 `word_tokenize` 進行分詞。

8. **移除停用詞 (Stopwords)**：移除英文常見停用詞如 "the", "is", "at" 等。

9. **詞形還原 (Lemmatization)**：使用 `WordNetLemmatizer` 將單詞還原為原型，如 "running" → "run"。

10. **移除短詞**：過濾掉長度小於 3 的單詞。

```python
# 預處理函數示例
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#(\w+)', r'\1', text)
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(token) for token in tokens 
              if token not in stop_words and len(token) > 2]
    return ' '.join(tokens)
```

````python
# 讀取訓練集的情緒標籤
df_emotion = pd.read_csv('/content/drive/MyDrive/ISA5810/dm-lab-2-private-competition/emotion.csv')
# 讀取訓練/測試集標示
df_identification = pd.read_csv('/content/drive/MyDrive/ISA5810/dm-lab-2-private-competition/data_identification.csv')
# 讀取提交格式範例
df_sample = pd.read_csv('/content/drive/MyDrive/ISA5810/dm-lab-2-private-competition/samplesubmission.csv')

# 合併所有數據
df = df_posts.merge(df_identification, on='id', how='left')
df = df.merge(df_emotion, on='id', how='left')

# 分割訓練集和測試集
train_df = df[df['split'] == 'train'].copy()
test_df = df[df['split'] == 'test'].copy()

# 訓練集有標籤，測試集的標籤是 NaN
print(f"訓練集大小: {len(train_df)}")
print(f"測試集大小: {len(test_df)}")
print(f"情緒類別: {train_df['emotion'].unique()}")
````

````python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    # 轉換為小寫
    text = text.lower()

    # 移除 URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # 移除 @用戶名 (mentions)
    text = re.sub(r'@\w+', '', text)

    # 移除 #hashtag 符號但保留文字
    text = re.sub(r'#(\w+)', r'\1', text)

    # 移除 HTML 標籤
    text = re.sub(r'<.*?>', '', text)

    # 移除表情符號和特殊字符（保留基本標點）
    text = re.sub(r'[^\w\s]', ' ', text)

    # 移除數字
    text = re.sub(r'\d+', '', text)

    # 移除多餘空白
    text = re.sub(r'\s+', ' ', text).strip()

    # 分詞
    tokens = word_tokenize(text)

    # 移除停用詞並進行詞形還原
    tokens = [lemmatizer.lemmatize(token) for token in tokens
              if token not in stop_words and len(token) > 2]

    return ' '.join(tokens)

train_df['cleaned_text'] = train_df['text'].apply(preprocess_text)
test_df['cleaned_text'] = test_df['text'].apply(preprocess_text)

# 顯示預處理結果
print("\n預處理前後對比:")
print("="*50)
for i in range(3):
    print(f"\n原始文本: {train_df['text'].iloc[i][:100]}...")
    print(f"清理後: {train_df['cleaned_text'].iloc[i][:100]}...")
    print(f"情緒標籤: {train_df['emotion'].iloc[i]}")
    print("-"*50)

# 檢查預處理後的統計資訊
print(f"\n訓練集清理後的空文本數量: {(train_df['cleaned_text'] == '').sum()}")
print(f"測試集清理後的空文本數量: {(test_df['cleaned_text'] == '').sum()}")
````

## 2. Feature Engineering Steps

````python
### Add the code related to the feature engineering steps in cells inside this section
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
import numpy as np

# ============================
# 1. TF-IDF 特徵提取
# ============================

# 建立 TF-IDF 向量化器
tfidf_vectorizer = TfidfVectorizer(
    max_features=5000,      # 最多保留 5000 個特徵詞
    ngram_range=(1, 2),     # 使用 unigram 和 bigram
    min_df=2,               # 最少出現在 2 個文檔中
    max_df=0.95,            # 最多出現在 95% 的文檔中
    sublinear_tf=True       # 使用 sublinear TF scaling
)

# 對訓練集擬合並轉換
X_train_tfidf = tfidf_vectorizer.fit_transform(train_df['cleaned_text'])

# 對測試集轉換（使用相同的向量化器）
X_test_tfidf = tfidf_vectorizer.transform(test_df['cleaned_text'])

print(f"TF-IDF 訓練集特徵維度: {X_train_tfidf.shape}")
print(f"TF-IDF 測試集特徵維度: {X_test_tfidf.shape}")

# ============================
# 2. 標籤編碼
# ============================

label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train_df['emotion'])

# 顯示標籤對應關係
print("\n標籤對應關係:")
for i, label in enumerate(label_encoder.classes_):
    print(f"  {i}: {label}")

# ============================
# 3. 額外文本特徵
# ============================

def extract_text_features(df, text_col='text', cleaned_col='cleaned_text'):
    features = pd.DataFrame()

    # 原始文本長度
    features['text_length'] = df[text_col].apply(lambda x: len(str(x)))

    # 清理後文本長度
    features['cleaned_length'] = df[cleaned_col].apply(lambda x: len(str(x)))

    # 單詞數量
    features['word_count'] = df[cleaned_col].apply(lambda x: len(str(x).split()))

    # 平均單詞長度
    features['avg_word_length'] = df[cleaned_col].apply(
        lambda x: np.mean([len(word) for word in str(x).split()]) if len(str(x).split()) > 0 else 0
    )

    # 驚嘆號數量（情緒指標）
    features['exclamation_count'] = df[text_col].apply(lambda x: str(x).count('!'))

    # 問號數量
    features['question_count'] = df[text_col].apply(lambda x: str(x).count('?'))

    # 大寫字母比例（情緒強度指標）
    features['caps_ratio'] = df[text_col].apply(
        lambda x: sum(1 for c in str(x) if c.isupper()) / max(len(str(x)), 1)
    )

    return features

# 提取額外特徵
train_extra_features = extract_text_features(train_df)
test_extra_features = extract_text_features(test_df)

print(f"\n額外特徵數量: {train_extra_features.shape[1]}")
print(f"額外特徵欄位: {list(train_extra_features.columns)}")

# ============================
# 4. 合併所有特徵
# ============================
from scipy.sparse import hstack

# 將額外特徵轉換為稀疏矩陣並與 TF-IDF 合併
X_train_combined = hstack([X_train_tfidf, train_extra_features.values])
X_test_combined = hstack([X_test_tfidf, test_extra_features.values])

print(f"\n合併後訓練集特徵維度: {X_train_combined.shape}")
print(f"合併後測試集特徵維度: {X_test_combined.shape}")

# ============================
# 5. 檢視特徵重要性（Top TF-IDF 詞彙）
# ============================
feature_names = tfidf_vectorizer.get_feature_names_out()
print(f"\n前20個 TF-IDF 特徵詞: {feature_names[:20]}")
````

````python
# ============================
# 進階特徵工程
# ============================
import re
import numpy as np

# 1. 情緒詞彙特徵字典
emotion_lexicon = {
    'joy': ['happy', 'joy', 'love', 'wonderful', 'amazing', 'great', 'excited', 'awesome', 'fantastic', 'blessed', 'grateful', 'delighted', 'cheerful', 'thrilled', 'ecstatic'],
    'anger': ['angry', 'hate', 'furious', 'mad', 'annoyed', 'frustrated', 'rage', 'irritated', 'pissed', 'outraged', 'hostile', 'bitter', 'resentful', 'infuriated'],
    'sadness': ['sad', 'depressed', 'unhappy', 'crying', 'miserable', 'heartbroken', 'lonely', 'disappointed', 'grief', 'sorrow', 'hopeless', 'devastated', 'gloomy', 'melancholy'],
    'fear': ['scared', 'afraid', 'terrified', 'anxious', 'worried', 'nervous', 'panic', 'horror', 'frightened', 'dread', 'alarmed', 'uneasy', 'paranoid', 'phobia'],
    'surprise': ['surprised', 'shocked', 'amazed', 'astonished', 'unexpected', 'stunning', 'unbelievable', 'wow', 'omg', 'whoa', 'speechless', 'startled', 'astounded'],
    'disgust': ['disgusted', 'gross', 'nasty', 'sick', 'horrible', 'awful', 'repulsive', 'revolting', 'yuck', 'ugh', 'repugnant', 'loathsome', 'vile', 'nauseating']
}

def count_emotion_words(text, emotion_words):
    """計算文本中特定情緒詞彙的數量"""
    if not isinstance(text, str):
        return 0
    text_lower = text.lower()
    return sum(1 for word in emotion_words if word in text_lower)

def extract_advanced_features(df, text_col='text', cleaned_col='cleaned_text'):
    """提取進階文本特徵"""
    features = pd.DataFrame()

    # ============================
    # 基本文本統計特徵
    # ============================
    features['text_length'] = df[text_col].apply(lambda x: len(str(x)))
    features['cleaned_length'] = df[cleaned_col].apply(lambda x: len(str(x)))
    features['word_count'] = df[cleaned_col].apply(lambda x: len(str(x).split()))
    features['avg_word_length'] = df[cleaned_col].apply(
        lambda x: np.mean([len(word) for word in str(x).split()]) if len(str(x).split()) > 0 else 0
    )
    features['unique_word_ratio'] = df[cleaned_col].apply(
        lambda x: len(set(str(x).split())) / max(len(str(x).split()), 1)
    )

    # ============================
    # 標點符號特徵（情緒強度指標）
    # ============================
    features['exclamation_count'] = df[text_col].apply(lambda x: str(x).count('!'))
    features['question_count'] = df[text_col].apply(lambda x: str(x).count('?'))
    features['ellipsis_count'] = df[text_col].apply(lambda x: str(x).count('...'))
    features['caps_ratio'] = df[text_col].apply(
        lambda x: sum(1 for c in str(x) if c.isupper()) / max(len(str(x)), 1)
    )
    features['caps_word_count'] = df[text_col].apply(
        lambda x: sum(1 for word in str(x).split() if word.isupper() and len(word) > 1)
    )

    # ============================
    # 情緒詞彙計數特徵
    # ============================
    for emotion, words in emotion_lexicon.items():
        features[f'{emotion}_word_count'] = df[text_col].apply(
            lambda x: count_emotion_words(x, words)
        )

    # ============================
    # 表情符號與社群媒體特徵
    # ============================
    features['emoji_positive'] = df[text_col].apply(
        lambda x: len(re.findall(r'[:;]=?[)D]|<3|\u263a|\u2764', str(x)))
    )
    features['emoji_negative'] = df[text_col].apply(
        lambda x: len(re.findall(r'[:;]=?[(]|D:', str(x)))
    )
    features['hashtag_count'] = df[text_col].apply(lambda x: str(x).count('#'))
    features['mention_count'] = df[text_col].apply(lambda x: str(x).count('@'))

    # ============================
    # 句子結構特徵
    # ============================
    features['sentence_count'] = df[text_col].apply(
        lambda x: len(re.split(r'[.!?]+', str(x)))
    )
    features['avg_sentence_length'] = features['word_count'] / features['sentence_count'].replace(0, 1)

    # ============================
    # 重複字符特徵（情緒強調）
    # ============================
    features['repeated_chars'] = df[text_col].apply(
        lambda x: len(re.findall(r'(.)\1{2,}', str(x)))
    )

    return features

# 應用進階特徵工程
print("提取進階特徵...")
train_advanced_features = extract_advanced_features(train_df)
test_advanced_features = extract_advanced_features(test_df)

print(f"進階特徵數量: {train_advanced_features.shape[1]}")
print(f"特徵欄位: {list(train_advanced_features.columns)}")
print(f"\n訓練集特徵統計:")
print(train_advanced_features.describe())

# ============================
# 將進階特徵與 TF-IDF 合併
# ============================
from scipy.sparse import hstack

# 合併 TF-IDF + 基本特徵 + 進階特徵
X_train_all = hstack([X_train_tfidf, train_advanced_features.values])
X_test_all = hstack([X_test_tfidf, test_advanced_features.values])

print(f"\n合併所有特徵後:")
print(f"訓練集特徵維度: {X_train_all.shape}")
print(f"測試集特徵維度: {X_test_all.shape}")
````

## 3. Model Implementation Steps

````python
### Add the code related to the model implementation steps in cells inside this section
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, f1_score, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# 資料分割（驗證用）
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train_combined, y_train,
    test_size=0.2,
    random_state=42,
    stratify=y_train
)
print(f"訓練子集大小: {X_tr.shape[0]}")
print(f"驗證集大小: {X_val.shape[0]}")

# 模型比較
print("\n" + "="*50)
print("模型比較（使用驗證集）")
print("="*50)

# 注意：MultinomialNB 需要非負特徵，因此需要單獨處理
# 先用 TF-IDF 特徵測試 MultinomialNB
print("\n訓練 Multinomial Naive Bayes（僅使用 TF-IDF 特徵）...")
X_tr_tfidf, X_val_tfidf, _, _ = train_test_split(
    X_train_tfidf, y_train,
    test_size=0.2,
    random_state=42,
    stratify=y_train
)

mnb_model = MultinomialNB(alpha=1.0)  # alpha 是平滑參數
mnb_model.fit(X_tr_tfidf, y_tr)
y_pred_mnb = mnb_model.predict(X_val_tfidf)
f1_mnb = f1_score(y_val, y_pred_mnb, average='macro')
print(f"Multinomial Naive Bayes - Macro F1 Score: {f1_mnb:.4f}")

# 其他模型使用合併特徵
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, C=1.0, random_state=42),
    'Linear SVC': LinearSVC(max_iter=2000, C=1.0, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42)
}

# 初始化最佳模型追蹤（包含 MNB 結果）
best_model = mnb_model
best_score = f1_mnb
best_model_name = "Multinomial Naive Bayes"
use_tfidf_only = True  # 標記是否只用 TF-IDF

for name, model in models.items():
    print(f"\n訓練 {name}...")
    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_val)

    # 計算 Macro F1 Score（競賽評估指標）
    f1 = f1_score(y_val, y_pred, average='macro')
    print(f"{name} - Macro F1 Score: {f1:.4f}")

    if f1 > best_score:
        best_score = f1
        best_model = model
        best_model_name = name
        use_tfidf_only = False

print(f"\n最佳模型: {best_model_name} (F1: {best_score:.4f})")

# 最佳模型詳細評估
print("\n" + "="*50)
print(f"最佳模型詳細評估: {best_model_name}")
print("="*50)

if use_tfidf_only:
    y_val_pred = best_model.predict(X_val_tfidf)
else:
    y_val_pred = best_model.predict(X_val)
print("\n分類報告:")
print(classification_report(y_val, y_val_pred, target_names=label_encoder.classes_))

# 使用全部訓練資料重新訓練最佳模型
print("\n使用全部訓練資料重新訓練最佳模型...")

if best_model_name == 'Multinomial Naive Bayes':
    final_model = MultinomialNB(alpha=1.0)
    final_model.fit(X_train_tfidf, y_train)
    X_test_final = X_test_tfidf
elif best_model_name == 'Logistic Regression':
    final_model = LogisticRegression(max_iter=1000, C=1.0, random_state=42)
    final_model.fit(X_train_combined, y_train)
    X_test_final = X_test_combined
elif best_model_name == 'Linear SVC':
    final_model = LinearSVC(max_iter=2000, C=1.0, random_state=42)
    final_model.fit(X_train_combined, y_train)
    X_test_final = X_test_combined
elif best_model_name == 'Random Forest':
    final_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
    final_model.fit(X_train_combined, y_train)
    X_test_final = X_test_combined
else:
    final_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
    final_model.fit(X_train_combined, y_train)
    X_test_final = X_test_combined

print("模型訓練完成！")

# 對測試集進行預測
print("\n對測試集進行預測...")
y_test_pred = final_model.predict(X_test_final)

# 將預測結果轉換回標籤名稱
y_test_labels = label_encoder.inverse_transform(y_test_pred)

print(f"預測完成！共 {len(y_test_labels)} 筆預測")
print(f"預測類別分布:")
for emotion in label_encoder.classes_:
    count = sum(y_test_labels == emotion)
    print(f"  {emotion}: {count}")

# 生成提交檔案
submission = pd.DataFrame({
    'id': test_df['id'].values,
    'emotion': y_test_labels
})

# 儲存提交檔案
submission.to_csv('/content/drive/MyDrive/ISA5810/dm-lab-2-private-competition/submission.csv', index=False)
print("\n提交檔案已儲存為 'submission.csv'")
print(submission.head(10))
````

````python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Bidirectional, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, Callback
from sklearn.model_selection import train_test_split

# ============================
# 1. 文本序列化
# ============================
MAX_WORDS = 25000 # 保留最常出現的前 25000 個單字
MAX_LEN = 100 # 設定一句話的最大長度

tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token='<OOV>')
tokenizer.fit_on_texts(train_df['cleaned_text'])

X_train_seq = tokenizer.texts_to_sequences(train_df['cleaned_text'])
X_test_seq = tokenizer.texts_to_sequences(test_df['cleaned_text'])

X_train_pad = pad_sequences(X_train_seq, maxlen=MAX_LEN, padding='post', truncating='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=MAX_LEN, padding='post', truncating='post')

# ============================
# 2. 建立 BiLSTM 模型
# ============================
num_classes = len(label_encoder.classes_)

model = Sequential([
    Embedding(MAX_WORDS, 128, input_length=MAX_LEN),
    Bidirectional(LSTM(256, return_sequences=True)),
    Bidirectional(LSTM(128)),
    Dense(64, activation='relu'),
    Dense(num_classes, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ============================
# 3. 訓練模型
# ============================
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train_pad, y_train,
    test_size=0.2,
    random_state=42,
    stratify=y_train
)

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Custom Callback to log every N epochs
class CustomLogger(Callback):
    def __init__(self, epochs_per_log=10):
        super().__init__()
        self.epochs_per_log = epochs_per_log

    def on_epoch_end(self, epoch, logs=None):
        if (epoch + 1) % self.epochs_per_log == 0 or epoch == 0: # Log first epoch and every Nth
            print(f"Epoch {epoch + 1}/{self.params['epochs']} - "
                  f"loss: {logs['loss']:.4f} - accuracy: {logs['accuracy']:.4f} - "
                  f"val_loss: {logs['val_loss']:.4f} - val_accuracy: {logs['val_accuracy']:.4f}")

history = model.fit(
    X_tr, y_tr,
    epochs=100,
    batch_size=32,
    validation_data=(X_val, y_val),
    callbacks=[early_stop, CustomLogger(epochs_per_log=10)],
    verbose=0 # Set verbose to 0 to only show output from CustomLogger
)

# ============================
# 4. 預測與提交
# ============================
y_test_pred = model.predict(X_test_pad)
y_test_classes = y_test_pred.argmax(axis=1)
y_test_labels = label_encoder.inverse_transform(y_test_classes)

submission_lstm = pd.DataFrame({
    'id': test_df['id'].values,
    'emotion': y_test_labels
})
submission_lstm.to_csv('/content/drive/MyDrive/ISA5810/dm-lab-2-private-competition/submission_lstm.csv', index=False)
print("LSTM 提交檔案已儲存！")
````

````python
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from sklearn.metrics import f1_score, precision_recall_fscore_support
from datasets import Dataset

# Get emotion labels and create id2label and label2id mappings
# Assuming label_encoder is already defined and fitted from previous steps
num_labels = len(label_encoder.classes_)
id2label = {i: label for i, label in enumerate(label_encoder.classes_)}
label2id = {label: i for i, label in enumerate(label_encoder.classes_)}

print(f"Number of labels: {num_labels}")
print(f"id2label mapping: {id2label}")
print(f"label2id mapping: {label2id}")
````

````python
model_checkpoint = "distilroberta-base"
# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

# Load model
model = AutoModelForSequenceClassification.from_pretrained(
    model_checkpoint,
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id
)

print(f"Tokenizer loaded: {type(tokenizer).__name__}")
print(f"Model loaded: {type(model).__name__}")
print(f"Model's number of labels: {model.config.num_labels}")
````

````python
MAX_SEQ_LENGTH = 128 # Define a suitable maximum sequence length for the transformer

def tokenize_function(examples):
    return tokenizer(examples['cleaned_text'], truncation=True, padding='max_length', max_length=MAX_SEQ_LENGTH)

# Create Hugging Face Dataset objects
train_data_dict = {'cleaned_text': train_df['cleaned_text'].tolist(), 'labels': y_train.tolist()}
test_data_dict = {'cleaned_text': test_df['cleaned_text'].tolist()}

# Convert to Hugging Face Dataset
train_dataset = Dataset.from_dict(train_data_dict)
test_dataset = Dataset.from_dict(test_data_dict)

# Tokenize the datasets
tokenized_train_dataset = train_dataset.map(tokenize_function, batched=True)
tokenized_test_dataset = test_dataset.map(tokenize_function, batched=True)

print(f"Original train_df head:\n{train_df.head()}")
print(f"\nTokenized train dataset features: {tokenized_train_dataset.features}")
print(f"Tokenized test dataset features: {tokenized_test_dataset.features}")
print(f"\nFirst tokenized training example:\n{tokenized_train_dataset[0]}")
````

````python
def compute_metrics(p):
    predictions, labels = p
    predictions = np.argmax(predictions, axis=1)
    f1_macro = f1_score(labels, predictions, average='macro')
    return {
        'f1_macro': f1_macro,
        'accuracy': (predictions == labels).astype(np.float32).mean()
    }

print("Compute metrics function defined.")
````

````python
output_dir = "/content/drive/MyDrive/ISA5810/dm-lab-2-private-competition/results_transformer"

training_args = TrainingArguments(
    output_dir=output_dir,
    learning_rate=2e-5,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    num_train_epochs=10,
    weight_decay=0.001,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="f1_macro",
    report_to="none", # Disable WandB or other reporting for simplicity
    push_to_hub=False,
    logging_dir=f"{output_dir}/logs",
    logging_steps=100,
    seed=42
)

print("TrainingArguments configured.")
````

````python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,
    eval_dataset=tokenized_train_dataset.select(range(int(len(tokenized_train_dataset) * 0.1))), # Using a small subset of train for eval for now
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

print("Trainer initialized.")
````
